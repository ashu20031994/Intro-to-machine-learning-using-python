# Basic x versus y plot using matplotlib


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10, 10, 100) # gives 100 data points between -10 to 10 which are equally spaced
y = np.sin(x) #gives sin value for eaxh data point x
plt.plot(x, y, marker = 'x') #plt is used to plot x versus y in graph
plt.show() #show is used to show graph
```

```python
from sklearn.datasets import load_iris # import iris dataset from preinstalled sklearn datasets
iris_dataset = load_iris() #loading dataset
```

```python
from sklearn.model_selection import train_test_split 
#dividing dataste into training and test dataset 
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state = 0)
```

```python
#calculate shape of each train and test data-set
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
print("Predicted test data : {}".format(y_pred))
```

```python
print("test set score {:.2f} ".format(np.mean(y_pred == y_test)))
```

```python
print("test set score : ",format(knn.score(X_test, y_test)))
```
